```
#adding/updating column
df=df.withColumn('Price',df['Price']+100.0)
df.show()
```

```
#Dropping nan values with threshold as 3(atleast non nan columns to stay in datafram)
df.na.drop(how="any",thresh=3).show()
```

```
#NAN transformation- same process as in scikit learn
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=['Odometer (KM)'], 
    outputCols=["{}_imputed".format(c) for c in ['Odometer (KM)']]
    ).setStrategy("median")
# Add imputation cols to df
imputer.fit(df).transform(df).show()
```

```
#Filtering Rows
df.filter((df['Odometer (KM)']<=20000) & 
                  (df['Odometer (KM)']>=10000)).show()
```

```
### Grouped to find the total doors manufactured
df.groupBy('Make').sum('Doors').show()
```

```
#Problem - To Find Most Popular Car Colour
df.groupBy('Colour').count().show()
```

```
#AGGREGATE FUNCTION
df.agg({'Doors':'sum'}).show()
```
